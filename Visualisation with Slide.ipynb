```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1000, 1000)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim(-1000,1000)
ax.set_ylim(-50,50)
a=1
b=1
c=0
y = a * np.sin(b*x-c)
line, = ax.plot(x, y)


def update(a=widgets.IntSlider(min=-100,max=100,step=1,value=10), b=widgets.IntSlider(min=-100,max=100,step=1,value=10), c=widgets.IntSlider(min=-100,max=100,step=1,value=10)):
    line.set_ydata(a * np.sin(b * x -c))
    fig.canvas.draw()

interact(update);
```
